```python
# Define the format_prompt function
def format_prompt(system_instruction, sample):
    dialogue = "\n".join([f"{i+1}. {line}" for i, line in enumerate(sample["dialogue"])])
    annotations = "\n".join([f"{i+1}. {line.split(': ')[0]}: {tag}" for i, (line, tag) in enumerate(zip(sample["dialogue"], sample["annotations"]))])
    return f"{system_instruction}\n\nDialogue:\n{dialogue}\n\nAnnotations:\n{annotations}"

# Define the request function
def request(samples, system_instruction):
    for sample in samples:
        messages = format_prompt(system_instruction, sample)
        response = openai.Completion.create(
            model="text-davinci-003",
            prompt=messages,
            max_tokens=50,
            temperature=0
        )
        annotations = response.choices[0].text.strip()
        print("Annotations:")
        print(annotations)

# Call the request function
request(samples, system_instruction)
```
